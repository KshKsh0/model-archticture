```python
from yolov1 import yolov1
```

```python
Yol=yolov1()
model=Yol._build_model()
```

```python
model.summary()
```

```python
import tensorflow as tf
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)            
model.compile(optimizer=optimizer, loss=Yol.yolo_loss)
```

```python
model.fit()
```

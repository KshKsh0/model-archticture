```python
import re

pattern = r"the (.*)er they were, the \1er they will be"
text = "the wiser they were, the wiser they will be"

match = re.search(pattern, text)
if match:
    print("Matched:", match.group(0))
else:
    print("No match")
```

```python
import re

text = r"some ppl loves do ppl"
pattern = r"(?:some |few) (ppl | cat) loves do \1 "

match = re.search(pattern, text)
if match:
    print("Matched:", match.group(0))
else:
    print("No match")
```

```python
# Capturing Group
pattern_capturing = r"(some|a few) (people|cats)"
text = "some people like some people"
match_capturing = re.search(pattern_capturing, text)

if match_capturing:
    print("Captured:", match_capturing.group(1), match_capturing.group(2))  # outputs 'some' and 'people'

# Non-Capturing Group
pattern_non_capturing = r"(?:some|a few) (people|cats)"
match_non_capturing = re.search(pattern_non_capturing, text)

if match_non_capturing:
    print("Captured:", match_non_capturing.group(1))  # only captures 'people' (no group for 'some')
```

```python
import re

# Regex pattern
pattern = r"(?:some|a few) (people|cats) like some \1"

# Test strings
test_strings = [
    "some people like some people",  # Match
    "a few cats like some cats",     # Match
    "some people like some cats",    # No match
    "a few dogs like some dogs"      # No match
]

# Check each string
for text in test_strings:
    match = re.search(pattern, text)
    if match:
        print(f"Match: {text}")
    else:
        print(f"No match: {text}")
```

```python
#email match
text = 'ahmedalhmouz10@gmail.com'
pattern = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[a-zA-Z]{2,6}$'  # Adjusted regex
result = re.search(pattern, text)

if result:
    print("Match found:", result.group())
else:
    print("No match")
```
